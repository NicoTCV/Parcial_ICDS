```python
# Examen parcial de CDE
import pandas as pd

linkco2= "https://github.com/NicoTCV/Parcial_ICDS/raw/main/export_CO2.csv"
linkscore="https://github.com/NicoTCV/Parcial_ICDS/raw/main/rise-2016-database-for-download.xlsx" # score por tipo de energía
linkwp="https://github.com/NicoTCV/Parcial_ICDS/raw/main/6.1.%20Total_World_Production.xlsx" # producción minera

dataco2=pd.read_csv(linkco2,sep=';')
datascore=pd.read_excel(linkscore)
datawp=pd.read_excel(linkwp)
```

```python
#                     Trabajaré con dataco2
# solo quiero las columnas name, value y date_of_information
```

```python
# borro las columnas que no deseo, y guardo cambios
dataco2.drop(columns=['slug','ranking','region','date_of_information'],inplace=True)
```

```python
# cambio el nombre a las columnas
dataco2.rename(columns={'name':'Country','value':'CO2_en_2019'},inplace=True)
dataco2
```

```python
#                  Trabajaré con datascore
# solo quiero las columnas Country, Renovable Energy score
# y el indicador 7: Carbon Pricing and Monitoring Mechanism
```

```python
# como lo que borraré es mucho... eligiré el método loc
```

```python
datascore=datascore.loc[:,['Country Categories','Unnamed: 5','Unnamed: 6']]
datascore
```

```python
# cambio el nombre a las columnas
datascore.rename(columns={'Country Categories':'Country','Unnamed: 5':'Renewable Energy score','Unnamed: 6':'Energy_Efficiency_score'},inplace=True)
# elimino la primera fila
import numpy as np
datascore.loc[0]=np.nan
datascore
```

```python
#          Trabajaré con datapm (producción minera)
```

```python
# primero cambiaré el nombre de las columnas
datawp.rename(columns={'6.1.8.1. Total Minerals Production 2021 by Country in metr. t; column A Countries, column B Total Production inclusive Bauxite, column C Total Production of Iron and Ferro-Alloy Metals, column D Total Production of Non-Ferrous Metals, column E Total Production of Precious Metals (tons), column F Total Production of Industrial Minerals inclusive Diamond, column G Total Production of Mineral Fuels.':'Country','Unnamed: 6':'Mineral Fuels'},inplace=True)
datawp
```

```python
# Df con nuevo contenido
datawp=datawp.loc[:,['Country','Mineral Fuels']]
# cambio nombre de columnas
datawp
```

```python
#  Ahora uniré las 3 tablas con merge
```

```python
dataco2.shape,datascore.shape,datawp.shape
```

```python
# usaré el método de fusión estándar
```

```python
merge1=dataco2.merge(datascore,how='outer',left_on='Country',right_on='Country')
merge1
```

```python
merge2=merge1.merge(datawp,how='left',left_on='Country',right_on='Country')
# cambiaré a mayúsculas la data merge2
merge2['Country']=merge2['Country'].str.upper()
merge2
# merge 2 es la unión de mis 3 datas
```

```python
# agrego data que mandó el profesor
```

```python
linkteach='https://github.com/NicoTCV/Parcial_ICDS/raw/main/country_isos.xlsx'

datateach=pd.read_excel(linkteach)
datateach
```

```python
# Voy a capitalizar los nombres de Country xq los datos del profesor están en mayúsculas
merge2['Country']=merge2['Country'].str.upper()
merge2
```

```python
# Ahora uniré esta data con la anterior, 
# priorizando las filas de la data del profe
```

```python
data_merge=datateach.merge(merge2,how='left',left_on='Country',right_on='Country')
data_merge
```

```python
# Modificaré el nombre de las columnas
# Official_State_Name, Internet_cc_TLD, iso_2,iso_3, Renewable_Energy_score,Mineral_Fuels
data_merge.rename(columns={'Officialstatename':'Official_State_Name','InternetccTLD':'Internet_CC_TLD','iso2':'ISO_2','iso3':'ISO_3','Renewable Energy score':'Renewable_Energy_score','Mineral Fuels':'Mineral_Fuels'},inplace=True)
data_merge
```

```python
# voy a eliminar espacios de las columnas
data_merge=data_merge.rename(columns=lambda x:x.strip())
```

```python
# voy a eliminar espacios antes y después del texto contenido en la columna 'Official_State_Name' y 'Country'
data_merge['Official_State_Name']=data_merge['Official_State_Name'].str.strip()
data_merge['Country']=data_merge['Country'].str.strip()
```

```python
# reemplazaré espacios de la columna "Official_State_Name"
data_merge['Official_State_Name']=data_merge['Official_State_Name'].str.replace(' ','_',regex=True)
data_merge
```

```python
# Pregunto si todos son dígitos 
data_merge.CO2_en_2019.str.contains(pat=r'\d',regex=True)
```

```python
# quiero convertir mis valores "numéricos" en eso, números
# antes corrijo los NaN en 0 para que lo reconozca
data_merge.replace(np.nan,0,inplace=True)
data_merge
```

```python
# elimino las , de 'CO2'
data_merge['CO2_en_2019']=data_merge['CO2_en_2019'].str.replace(',','')
```
